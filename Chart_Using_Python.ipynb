```python
import matplotlib.pyplot as plt
```

```python
# Data
parameters = [
    "Locule damage",
    "Yield Loss",
    "Seed cotton yield loss",
    "Oil content loss",
    "Boll opening loss"
]

col1 = [0,35,3,2,11]
col2 = [55,90,62,47,59]

y_positions = range(len(parameters))

plt.figure(figsize=(10,5))

# Scatter Points
plt.scatter(col1, y_positions)
plt.scatter(col2, y_positions)

# Connecting each pair of points with a line
for y, x1, x2 in zip(y_positions, col1, col2):
  plt.plot([x1,x2], [y,y])

# Adding data label
for y, x in zip(y_positions, col1):
  plt.text(x-1, y, f"{x}%", va = "center", ha = "right")

for y, x in zip(y_positions, col2):
  plt.text(x+1, y, f"{x}%", va = "center", ha = "left")

# Formatting
plt.yticks(y_positions, parameters)
plt.xlabel("Percentage")
plt.tight_layout()
plt.savefig("chart.png", dpi= 300, bbox_inches = "tight")
plt.show()
```
